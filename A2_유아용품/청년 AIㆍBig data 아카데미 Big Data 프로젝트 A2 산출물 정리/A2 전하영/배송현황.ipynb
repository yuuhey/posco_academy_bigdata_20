```python
df_area = pd.merge(df_M, df_S, on = ['고객번호'], how = 'left')
df_area
```

```python
df_area.isnull().sum()
```

```python
df_area.dropna(subset = ['거주지역'])
```

```python
df_area['구매일'] = pd.to_datetime(df_area['구매일'])
```

```python
df_area['배송시작일'] = pd.to_datetime(df_area['배송시작일'])
```

```python
df_area['배송완료일'] = pd.to_datetime(df_area['배송완료일'])
```

```python
df_area['거주지역'] = df_area['거주지역'].astype(str)
```

```python
df_area.info()
```

```python
df_area['배송기간'] = df_area['배송완료일'] - df_area['구매일']
```

```python
df_area['출고기간'] = df_area['배송완료일'] - df_area['배송시작일']
```

```python
df_area['배송기간'] = df_area['배송기간'].dt.days
```

```python
df_area['출고기간'] = df_area['출고기간'].dt.days
```

```python
df_area['거주지역'] = df_area['거주지역'].astype(str)
```
