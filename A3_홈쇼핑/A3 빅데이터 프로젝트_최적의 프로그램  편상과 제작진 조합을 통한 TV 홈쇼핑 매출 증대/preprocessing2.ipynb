#### 공휴일 및 이벤트날에 대한 변수 생성 weekend

```python
weekend = ['200124', '200125', '200126', '200127', '200301'
           '200415', '200430', '200505', '200606', '200815',
           '200930', '201001', '201002', '201003', '201009',
           '201225', '210101', '210211', '210212', '210213',
           '210301', '210505', '210519', '210606', '210815',
           '210816', '210920', '210921', '210922', '211003',
           '211004', '211009', '211011']
```

```python
def is_weekend(x):
    if x == 5 or x == 6: 
        return 1
    else: return 0
data1['weekend'] = data1['요일'].apply(lambda x:is_weekend(x))
data1['weekend']
```

```python
data1['weekend'].value_counts()
```

```python
data1[data1['weekend'] == 0]['date'].value_counts()
```

```python
days = data1[data1['weekend'] == 0]
days
```

```python
def is_holidays(x):
    weekend = ['2020-01-24', '2020-01-25', '2020-01-26', '2020-01-27', '2020-03-01'
           '2020-04-15', '2020-04-30', '2020-05-05', '2020-06-06', '2020-08-15',
           '2020-09-30', '2020-10-01', '2020-10-02', '2020-10-03', '2020-10-09',
           '2020-12-25', '2021-01-01', '2021-02-11', '2021-02-12', '2021-02-13',
           '2021-03-01', '2021-05-05', '2021-05-19', '2021-06-06', '2021-08-15',
           '2021-08-16', '2021-09-20', '2021-09-21', '2021-09-22', '2021-10-03',
           '2021-10-04', '2021-10-09', '2021-10-11']
    if x in weekend:
        return 1
    else:
        return 0
data1['holi'] = data1['date'].apply(lambda x: is_holidays(x))
data1['holi'].value_counts()
```

```python
day_idx = days.index
data1.loc[day_idx,'weekend'] = data1.loc[day_idx,'weekend'] + data1.loc[day_idx,'holi']
```
