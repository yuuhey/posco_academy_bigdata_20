'Large Lymphocyte' : 림프구 -> 림프계, 혈액, 골수에 존재하며 B세포와 T세포로 분류되어 정상적으로는 세균, 바이러스나 질병을 일으키는 물질과 싸워 감염과 질병으로부터 우리 몸을 보호하는 역할을 맡고 있다.

'Location of herniation' : -> 디스크 탈출 위치 

'OswestryDiasbillityIndex(ODI)' : 요통 기능 장애 평가

'가족력' : 유전적 질병을 얘기 하는 것이 아닌, 가족중 1명이라도 디스크 환자가 있다는 것으 얘기하는 것이다.

'간질성폐질환' : 매우 어려운 얘기이나 증상으로는 '가장 특징적인 증상은 숨이 찬 것입다.' 아마 운동에 영향을 줄것이다.

'고혈압여부' :  혈압이 여러 원인으로 인해 높아진 상태 -> 운동을 하는데 영향을 줄것으로 예상

'과거수술횟수' : 디스크 과거 수술 여부

'당뇨여부' : 당뇨와 만성 목, 허리 통증 발생은 밀접한 관련성이 있으며 당뇨병이 이러한 근골격계 문제를 발생시키는 위험요소 중 하나라고 보고하였다(목 통증 27.3%, 허리 통증 34.%).
            당뇨병과 척추의 병리학적 변화는 당뇨병에서 일반적으로 나타나는 지방 대사의 변화와 고혈당의 결과와 관련이 있다. 즉, 이러한 변화가 척추의 디스크에 영향을 주어 척추의 병리적 원인이 될 수 있다는 것이다.
            
'말초동맥질환여부' :  다리에 뻗어 있는 동맥에 혈전·지방 등이 달라붙어 혈관이 좁아지는 질환이다. /하지만 비율은 5%미만으로 매우 낮은 수치를 보여주며 말초동맥질환이 발병하고 디스크라고 착각 할 수 있다고 판단되는 경우가 있다.

'빈혈여부','헤모글로빈수치' : 혈액 내 적혈구 감소로 산소 공급이 원활하지 않아 어지럼증이 나타나는 질환 -> 운동에 영향이 있을 것이라고 판단.

'성별' : 나이가 어리면 동일한 디스크 비율을 보여주지만 나이가 들고 50대의 디스크 비율을 살펴보면 여자가 남자보다 더 높은 비율로 디스크환자가 있다는 것을 확인 할 수 있다.

'스테로이드치료' : 염증이 생긴 신경 주위에 스테로이드, 국소마취제 등의 약물을 투입하여 염증을 제거하고 통증 전달을 차단시키는 치료. 허리 주사치료에 사용되는 약물

'신부전여부' :  신장질환으로 운동을 해야되는 요소라 빼도 된다고 생각해 드롭

'신장(키)' : 키가 크다고 디스크에 걸릴확률이 높다는 사실은 입증된 사실이 없다. 즉 키가 디스크에 영향을 주는 것은 아니기에 제외해도 된다고 생각함

'심혈관질환' : 심장병의 종류로 디스크에 직접적인 영향은 없으나 운동 시 주의해야하는 사항이 있으므로 운동쪽이랑 엮으면 좋을것 같다.

'암발병여부' 제외

'연령': 나이는 들면 들수록 디스크환자일 확률이 증가하는 것을 확인 했다. 하여 연령은 나이와 관계가 있으므로 디스크와 연관시키면 좋을 것이다.

'우울증여부' : 디스크 질환과 관게가 없다.

'입원기간','입원일자','퇴원일자' : 수술과 관계가 있을 것이니 내비둠

'종양진행여부' : 디스크와 연관 없음

'직업' : 직업에 따라 디스크 여부가 나뒨다. 사무직과 현장직을 예로 들면 사무직은 장시간 의자에 오래앉아 있기 때문에 디스크 발병 확률이 높다.

'체중': 몸무게가 많이 나갈수 록 허리에 가해지는 영향이 크기 때문에 체중의 영향이 크다고 판단.

'혈전합병증여부' :정맥에 생기는 혈전합병증, 디스크와 관련이 없다.

'환자통증정도' : 통증의 정도가 높을수 록 디스크 정도를 파악 할 수 있다고 생각해 버리지 않음

'흡연여부' :  흡연여부에 따른 디스크와 관련성은 크다고 보지 않으므로 일단 분석 후 제외를 할지 고민한다.

'통증기간(월)' : 통증기간은 이렇게 생각했다. 허리디스크라고 판단이 되어도 바로 수술을 하는 것이 아닌, 비수술 치료 기간을 가진 뒤, 비수술 치료로 해결이 되지 않는 다면, 수술 치료로 넘어가는 것이다.

# K-means

```python
# 최대 군집 생성 개수
n_iter_cluster = 15
cluster_range = [i+1 for i in range(n_iter_cluster)]
clus_error = []
for v_n_clus in cluster_range:
    clus = KMeans(v_n_clus)
    clus.fit(df_raw_x)
    
    # 각 데이터로부터 가장 가까운 군집 중심점까지 거리 제곱합
    clus_error.append(clus.inertia_)
ds_error = pd.DataFrame({"NumberofCluster": cluster_range, "Error": clus_error})
ds_error.round(3)
```

군집 개수가 증가함에 따라 중심점까지의 거리는 가까워짐 ERROR는 거리제곱을 뜻함

```python
plt.figure(figsize=(10,10))
plt.plot(ds_error["NumberofCluster"], ds_error["Error"])
plt.title("Sum of squared distance")
plt.xlabel("Clusters")
plt.ylabel("Sum of squared distance")
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from sklearn.metrics import precision_score, recall_score, f1_score
# 변수 중요도 확인: F1 점수 기준
from xgboost import plot_importance
```

```python
# 분류모델 통합 평가: 혼동행렬, 정확도, 정밀도, 재현율, F1, AUC 등
def eval_class_model(y_test, y_pred):
    confusion = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    F1 = f1_score(y_test, y_pred)
    AUC = roc_auc_score(y_test, y_pred)
    
    print('오차행렬:\n', confusion, '\n')
    print('정확도: {:.4f}'.format(accuracy))
    print('정밀도: {:.4f}'.format(precision))
    print('재현율: {:.4f}'.format(recall))
    print('F1    : {:.4f}'.format(F1))
    print('AUC   : {:.4f}'.format(AUC))
```

```python
# 기본 모델 지정 및 기본 옵션 확인
xgboost_uncust = XGBClassifier()

xgboost_uncust
```

```python
# 모델 적합
xgboost_uncust.fit(df_train_x, df_train_y)

print('\n 모델 생성 및 평가 : Train \n')
# 예측 및 모델 평가:train  
xgb_pred_train = xgboost_uncust.predict(df_train_x)
# 모델 평가
eval_class_model(df_train_y, xgb_pred_train)

print('\n 모델 생성 및 평가 : Test \n')
# 예측 및 모델 평가:test 
xgb_pred_test = xgboost_uncust.predict(df_test_x)
# 모델 평가
eval_class_model(df_test_y, xgb_pred_test)
```

```python
# 사용자 모델 생성
xgboost_user= XGBClassifier(n_estimators = 100, learning_rate = 0.1, max_depth = 5)
# 모델 적합
xgboost_user.fit(df_train_x, df_train_y)

# 예측 및 모델 평가:Train  
xgb_pred_train = xgboost_user.predict(df_train_x)

print('\n 사용자 모델 평가 : Train \n')
eval_class_model(df_train_y, xgb_pred_train)
print('\n',classification_report(df_train_y, xgb_pred_train))


# 예측 및 모델 평가:Test  
xgb_pred_test = xgboost_user.predict(df_test_x)

print('\n 사용자 모델 평가 : Test \n')
eval_class_model(df_test_y, xgb_pred_test)
print('\n',classification_report(df_test_y, xgb_pred_test))
```

```python
# sklearn XGBoost 모델의 조기 중단 : fit( )에 파라미터 입력(early_stopping_rounds, eval_metrics, eval_set.)
# max_depth, learning_rate(학습률), 반복횟수 등 지정
# 오류함수의 평가지표:logloss
# 조기중단을 위한 최소 반복횟수는 150

xgb_stop = XGBClassifier(n_estimators = 300, learning_rate = 0.1 , max_depth = 6)

eval_df = [(df_test_x, df_test_y)]

xgb_stop.fit(df_train_x, df_train_y, early_stopping_rounds = 150, 
                eval_metric="logloss", eval_set = eval_df, verbose=True)
```

```python
# 조기 중단 예측 및 모델 평가:Test  
xgb_pred_stop = xgb_stop.predict(df_test_x)

print('\n 조기 중단 모델 평가 : Test \n')
eval_class_model(df_test_y, xgb_pred_stop)
print('\n',classification_report(df_test_y, xgb_pred_stop))
```
